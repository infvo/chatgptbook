```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Definiren van de functie en rotatie
def solid_of_revolution(y, max_y):
    return 2 * y

y = np.linspace(0, 2, 400)
x = solid_of_revolution(y, 2)

theta = np.linspace(0, 2 * np.pi, 400)
Y, Theta = np.meshgrid(y, theta)
X = solid_of_revolution(Y, 2) * np.cos(Theta)
Z = solid_of_revolution(Y, 2) * np.sin(Theta)

# Plotten van de figuur
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Oppervlak tekenen
ax.plot_surface(X, Y, Z, alpha=0.7, cmap='viridis')

# Aslabels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Titel
ax.set_title('Solid of Revolution of the Area between y=2, y=0.5x, and x-axis around the y-axis')

plt.show()
```
